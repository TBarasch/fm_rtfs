```python
# functions
def helper_value(value: str) -> int:
    d = {'K':'000', 'M': '000000', 'B':'000000000'}
    for letter in d.keys():
        if letter in value:
            value = value.replace(letter,'') + d[letter]
    if "." in value:
        value = value.replace('.','')[:-1]
    return int(value)
```

```python
df = pd.read_csv("sample_rtf.rtf", sep="|")
df = df[df.index % 2 != 0]
df.columns = [i.lstrip().rstrip().replace(' ', '_') for i in df.columns]
df = df.applymap(lambda x: x.lstrip().rstrip() if isinstance(x, str) else x)
```

```python
# dealing with hight column
df[['Height','Height_unit']] = pd.DataFrame(df.Height.str.split(' ').values.tolist())
# dealing with weight column
df[['Weight','Weight_unit']] = pd.DataFrame(df.Weight.str.split(' ').values.tolist())
# dealing with transfer value column
df[['Trasfer_Value_Lower', 'Transfer_Value_Upper']] = pd.DataFrame(df['Transfer_Value'].str.findall(r'[\d.KMB]+').apply(lambda x: x if len(x) == 2 else x*2).values.tolist()).applymap(helper_value)
```
